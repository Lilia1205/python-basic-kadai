```python
#5章課題
top_base=10
bottom_base=20
height=5
print(f"{(top_base+bottom_base)*height/2}㎠")
```

```python
#小数点を消したい
top_base=10
bottom_base=20
height=5
result = int((top_base+bottom_base)*height/2) # int()で小数点以下を切り捨て ます。
print(f"{result:}㎠")
```

```python
top_base=10
bottom_base=20
height=5
result =(top_base+bottom_base)*height/2 #Pythonの f文字列 (f"{result:.0f}") で小数点以下を表示させない方法
print(f"{result:.0f}㎠") #㎠が復活
```
